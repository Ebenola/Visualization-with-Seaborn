```python
# Creating a Bar Chart in Python using Seaborn
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Seaborn enhances data visualization

import seaborn as sns
sns.set()

# The sns.set() will enable the graphs to be made by matplotlib but displayed with seaborn outlook
```

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")

df_used_cars
```

```python
# Use Shift Tab after plt.bar to see the required details

# I increased the size of the plot so the x-axis labels don't overlap using plt.figure
# Alternatively, use plt.xticks to rotate the x-axis labels by 45 degrees so they dont overlap

plt.figure(figsize = (9, 6))

plt.xticks(rotation = 45, fontsize = 13)

plt.bar(x = df_used_cars["Brand"],
       height = df_used_cars["Cars Listings"], color = "midnightblue")
plt.yticks(fontsize = 13)
plt.title("Cars Listings by Brand", fontsize = 13, fontweight = "bold")
plt.ylabel("Number of Listings", fontsize = 13)

plt.show()

plt.savefig("Cars Listings Bar.png")
```
